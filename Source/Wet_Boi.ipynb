```python
pgh_wet_data = pd.read_csv("https://data.wprdc.org/datastore/dump/513290a6-2bac-4e41-8029-354cbda6a7b7",
                           index_col="id", 
                           parse_dates=True)
pgh_wet_data.head()
```

```python
wet_count = pgh_wet_data.groupby("neighborhood")["neighborhood"].count()
wet_count.sort_values(ascending=False).plot(x ='neighborhood', y='count', kind = 'bar', rot=90, figsize=(15,5))
```
